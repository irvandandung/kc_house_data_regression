```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df.drop(['date','id', 'zipcode','view','lat','long','sqft_living15', 'sqft_lot15'], axis=1, inplace=True)
df.head(2)
```

```python
f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['bedrooms'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['bedrooms'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['bathrooms'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['bathrooms'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['sqft_living'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['sqft_living'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['sqft_lot'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['sqft_lot'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['floors'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['floors'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['waterfront'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['waterfront'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['condition'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['condition'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['grade'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['grade'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['sqft_above'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['sqft_above'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['sqft_basement'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['sqft_basement'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['yr_built'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['yr_built'])
plt.show()

f = plt.figure(figsize=(12,4))
f.add_subplot(1,2,1)
df['yr_renovated'].plot(kind='kde')
f.add_subplot(1,2,2)
plt.boxplot(df['yr_renovated'])
plt.show()
```
